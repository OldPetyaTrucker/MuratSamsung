```python
X = data[num_cols + cat_cols].copy()
y = data[target_col].copy()

binary_cols = [
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
    'IsBillingPaperless'
]

binary_mapping = {
    'Sex': {'Female': 0, 'Male': 1},
    'IsSeniorCitizen': {'No': 0, 'Yes': 1},
    'HasPartner': {'No': 0, 'Yes': 1},
    'HasChild': {'No': 0, 'Yes': 1},
    'HasPhoneService': {'No': 0, 'Yes': 1},
    'IsBillingPaperless': {'No': 0, 'Yes': 1}
}

for col in binary_cols:
    if col not in X.columns:
        raise ValueError(f"Колонка {col} отсутствует в данных")
    if X[col].dtype == 'object':
        X[col] = X[col].map(binary_mapping.get(col, {})).fillna(-1)
    X[col] = pd.Categorical(
        X[col],
        categories=[0, 1],
        ordered=False
    )
for col in binary_cols:
    print(f"\n{col}:")
    print("Уникальные значения:", X[col].unique())
    print("Распределение:")
    print(X[col].value_counts())
```

```python
import seaborn as sns
def check_total_spent(df, column='TotalSpent'):
    print(f"Тип данных: {df[column].dtype}")
    print(f"Всего записей: {len(df)}")
    print(f"Уникальных значений: {df[column].nunique()}")
    print(f"Количество пропусков (NaN): {df[column].isna().sum()}")
    non_numeric = df[column].apply(lambda x: not isinstance(x, (int, float)) and not pd.isna(x))
    if non_numeric.any():
        print(f"Найдено нечисловых значений: {non_numeric.sum()}")
    else:
        print("Нечисловых значений не обнаружено ✓")
    numeric_series = pd.to_numeric(df[column], errors='coerce')
    negative = (numeric_series < 0).sum()
    zero = (numeric_series == 0).sum()
    print(f"Отрицательных значений: {negative}")
    print(f"Нулевых значений: {zero}")
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(numeric_series.dropna(), bins=30, kde=True)
    plt.xlabel(column)
    plt.subplot(1, 2, 2)
    sns.boxplot(x=numeric_series.dropna())
    plt.xlabel(column)
    plt.tight_layout()
    plt.show()
check_total_spent(X)
```

```python
from catboost import CatBoostClassifier, Pool


X_train, X_val, y_train, y_val = train_test_split(
    data[num_cols + cat_cols],
    data[target_col],
    test_size=0.2,
    random_state=42,
    stratify=data[target_col]
)

train_pool = Pool(
    X_train,
    label=y_train,
    cat_features=cat_cols
)

val_pool = Pool(
    X_val,
    label=y_val,
    cat_features=cat_cols
)


base_model = CatBoostClassifier(
    random_seed=42,
    verbose=0,
    auto_class_weights='Balanced'
)

base_model.fit(train_pool)
base_pred = base_model.predict_proba(val_pool)[:, 1]
print(f"рогиауги: {roc_auc_score(y_val, base_pred):.4f}")


param_grid = {
    'iterations': [500, 1000, 1500],
    'learning_rate': [0.03, 0.1, 0.3],
    'depth': [4, 6, 8]
}

best_score = 0
best_params = {}

for n_estimators in param_grid['iterations']:
    for lr in param_grid['learning_rate']:
        for depth in param_grid['depth']:
            model = CatBoostClassifier(
                iterations=n_estimators,
                learning_rate=lr,
                depth=depth,
                random_seed=42,
                early_stopping_rounds=50,
                verbose=0,
                auto_class_weights='Balanced'
            )

            model.fit(
                train_pool,
                eval_set=val_pool,
                use_best_model=True
            )

            pred = model.predict_proba(val_pool)[:, 1]
            score = roc_auc_score(y_val, pred)

            if score > best_score:
                best_score = score
                best_params = {
                    'iterations': n_estimators,
                    'learning_rate': lr,
                    'depth': depth
                }

            print(f"{n_estimators}, {lr},{depth} | рогауг: {score:.4f}")

print(f"лучши параметры: {best_params}")
print(f"рог ауг скор лучшй: {best_score:.4f}")

final_model = CatBoostClassifier(
    **best_params,
    random_seed=42,
    early_stopping_rounds=50,
    verbose=0,
    auto_class_weights='Balanced'
)

final_model.fit(
    Pool(data[num_cols + cat_cols], data[target_col], cat_features=cat_cols),
    eval_set=val_pool
)
```
